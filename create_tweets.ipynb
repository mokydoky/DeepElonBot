```python
import openai
import numpy as np
prompt = ""
model = "ada:ft-personal-2023-01-03-03-07-23"
```

```python
arraycompletions = np.array([])
for i in np.arange(50):
    arraycompletions = np.append(arraycompletions, openai.Completion.create(model=model,prompt=prompt,max_tokens=50,stop="END"))
```

```python
for i in np.arange(50):
    print(arraycompletions[i]['choices'][0]['text'])
    print("SEPARATOR---------------------------")
```
